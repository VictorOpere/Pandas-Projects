```python
import pandas as pd

df = pd.read_csv("apple_july.csv",parse_dates=["Date"],index_col="Date")

df.head(10)
```

```python
%matplotlib inline 
df["2019-07-01":"2019-07-11"].plot()
```

```python
%matplotlib inline 
df["2019-07-01":"2019-07-11"].Close.plot()
```

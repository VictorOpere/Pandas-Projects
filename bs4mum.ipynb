```python
from bs4 import BeautifulSoup, SoupStrainer
import urllib.request as request

with request.urlopen('https://backbenchas.com/damascus-moments/') as response:
   html = response.read()
```

```python
only_body_tag = SoupStrainer('body')
soup = BeautifulSoup(html, 'html.parser', only_body_tag )


print(soup.prettify())

# print(dir(BeautifulSoup))

```

```python
print(soup.title)
```

```python
print(soup.title.name)
```

```python
print(soup.get_text())
```

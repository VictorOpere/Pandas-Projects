```python
# import beautifulsoup4
# import urllib 


from bs4 import BeautifulSoup, SoupStrainer
import urllib.request as request

```

```python
# making request to the website

with request.urlopen('https://backbenchas.com/damascus-moments/') as response:
    html = response.read()
    
    
    
```

```python
 only_body_content = SoupStrainer('body')


soup = BeautifulSoup(html, 'html.parser', parse_only=only_body_content)
```

```python
print(soup.prettify())
```

```python
# show title tag

soup.title
```

```python
# soup.title.text
```

```python
print(soup.get_text())
```

```python
# find all a tags

all_links = soup.find_all('a')

for link in all_links:
    print(link.get_text())
    
```

```python
the_content = soup.find("div", "post-content-body")
print(the_content)
```

```python
print(the_content.get_text())
```

```python
import pandas as pd
import numpy as np


df = pd.read_csv("weather_data3.csv")

df
```

```python
# replace data 

new_df = df.replace([-99999,-88888, 0],np.NaN)

new_df
 
```

```python
# replace data with specific column

new_df = df.replace({
    'temperature':-99999,
    'windspeed': -99999,
    'event': 0
},np.NaN)

new_df
```

```python

# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```
